### Path Setup + Fit Surrogates

```python
import sys
from pathlib import Path

repo_root = Path.cwd()
while repo_root.name != "Coulombic-Efficiency-Driven-Optimization-of-Health-Aware-Charging-Protocols":
    repo_root = repo_root.parent
sys.path.insert(0, str(repo_root))

from utils.soc_sweep_analysis import get_default_groups
from utils.surrogate_models import (
    collect_ce_surrogate_data,
    fit_poly2_linear,
    fit_gpr_surrogate,
)

processed_root = repo_root / "data" / "SOC sweep data (processed)"
groups = get_default_groups()

X, y, df = collect_ce_surrogate_data(processed_root, groups, ce_min=97, ce_max=100)
print(X.shape, y.shape)

poly, lin = fit_poly2_linear(X, y)
gpr = fit_gpr_surrogate(X, y)
print("GPR kernel:", gpr.kernel_)
```

### Linear Regression Surrogate Plots

```python
from utils.surrogate_models import plot_surrogate_fits_1x4, plot_surrogate_contours_1x4

plot_surrogate_fits_1x4(
    X,
    y,
    model=lin,
    poly=poly,
    # outpath="../Raw Exported Figures/Lin_surr_fits.svg",
)

plot_surrogate_contours_1x4(
    model=lin,
    poly=poly,
    # outpath="../Raw Exported Figures/Lin_surr_predicted_CE.svg",
)
```

### GPR Surrogate Plots

```python
plot_surrogate_fits_1x4(
    X,
    y,
    gpr=gpr,
    # outpath="../Raw Exported Figures/GP_surr_fits.svg",
)

plot_surrogate_contours_1x4(
    gpr=gpr,
    # outpath="../Raw Exported Figures/GPR_surr_predicted_CE.svg",
)
```
